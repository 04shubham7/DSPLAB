```python
import numpy as np
import matplotlib.pyplot as plt

# Input signal and impulse response
x = [1, 2, 1, 3, 1]
h = [1, 2, 1, 1, 1]

# Calculate the convolution of x and h
y = np.convolve(x, h)

# Create a figure with three subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))
fig.tight_layout(pad=3.0)

# Plot the input signal x(n)
ax1.stem(x)
ax1.set_title('Input Signal x(n)')
ax1.set_xlabel('n')
ax1.set_ylabel('x(n)')
ax1.grid(True)
ax1.set_xticks(np.arange(len(x))) # Set x-ticks to be integers

# Plot the impulse response h(n)
ax2.stem(h)
ax2.set_title('Impulse Response h(n)')
ax2.set_xlabel('n')
ax2.set_ylabel('h(n)')
ax2.grid(True)
ax2.set_xticks(np.arange(len(h))) # Set x-ticks to be integers


# Plot the output signal y(n)
ax3.stem(y)
ax3.set_title('Output Signal y(n) = x(n) * h(n)')
ax3.set_xlabel('n')
ax3.set_ylabel('y(n)')
ax3.grid(True)
ax3.set_xticks(np.arange(len(y))) # Set x-ticks to be integers


plt.show()
```
