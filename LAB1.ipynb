```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for continuous sine wave
x_continuous = np.linspace(-2*np.pi, 2*np.pi, 500)
sin_wave_continuous = np.sin(x_continuous)

# Plot continuous sine wave
plt.figure(figsize=(6, 4))
plt.plot(x_continuous, sin_wave_continuous, label='Continuous Sin Wave')
plt.title('Continuous Sine Wave')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete sine wave
x_discrete = np.linspace(-2*np.pi, 2*np.pi, 50)
sin_wave_discrete = np.sin(x_discrete)

# Plot discrete sine wave
plt.figure(figsize=(6, 4))
plt.stem(x_discrete, sin_wave_discrete, label='Discrete Sin Wave', basefmt=' ')
plt.title('Discrete Sine Wave')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for continuous e^x wave where x < 0
x_continuous_neg = np.linspace(-2*np.pi, -0.001, 250) # Exclude x=0
exp_wave_continuous_neg = np.exp(x_continuous_neg)

# Plot continuous e^x wave where x < 0
plt.figure(figsize=(6, 4))
plt.plot(x_continuous_neg, exp_wave_continuous_neg, label='e^x Wave (x < 0)')
plt.title('Continuous e^x Wave (x < 0)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for continuous e^x wave where x >= 0
x_continuous_pos = np.linspace(0, 2*np.pi, 250)
exp_wave_continuous_pos = np.exp(-x_continuous_pos) # Using e^-x for x>=0 as in previous plots

# Plot continuous e^x wave where x >= 0
plt.figure(figsize=(6, 4))
plt.plot(x_continuous_pos, exp_wave_continuous_pos, label='e^-x Wave (x >= 0)')
plt.title('Continuous e^-x Wave (x >= 0)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete e^x wave where x < 0
x_discrete_neg = np.linspace(-2*np.pi, -0.001, 25) # Exclude x=0
exp_wave_discrete_neg = np.exp(x_discrete_neg)

# Plot discrete e^x wave where x < 0
plt.figure(figsize=(6, 4))
plt.stem(x_discrete_neg, exp_wave_discrete_neg, label='e^x Wave (x < 0)', basefmt=' ')
plt.title('Discrete e^x Wave (x < 0)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete e^x wave where x >= 0
x_discrete_pos = np.linspace(0, 2*np.pi, 25)
exp_wave_discrete_pos = np.exp(-x_discrete_pos) # Using e^-x for x>=0 as in previous plots

# Plot discrete e^x wave where x >= 0
plt.figure(figsize=(6, 4))
plt.stem(x_discrete_pos, exp_wave_discrete_pos, label='e^-x Wave (x >= 0)', basefmt=' ')
plt.title('Discrete e^-x Wave (x >= 0)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for continuous e^x wave with both cases
x_continuous_neg = np.linspace(-2*np.pi, -0.001, 250) # Exclude x=0 for x<0
exp_wave_continuous_neg = np.exp(x_continuous_neg)

x_continuous_pos = np.linspace(0, 2*np.pi, 250)
exp_wave_continuous_pos = np.exp(-x_continuous_pos) # Using e^-x for x>=0 as in previous plots

# Plot continuous e^x wave with both cases
plt.figure(figsize=(6, 4))
plt.plot(x_continuous_neg, exp_wave_continuous_neg, label='e^x Wave (x < 0)')
plt.plot(x_continuous_pos, exp_wave_continuous_pos, label='e^-x Wave (x >= 0)')
plt.title('Continuous e^|x| Wave')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete e^x wave with both cases
x_discrete_neg = np.linspace(-2*np.pi, -0.001, 25) # Exclude x=0 for x<0
exp_wave_discrete_neg = np.exp(x_discrete_neg)

x_discrete_pos = np.linspace(0, 2*np.pi, 25)
exp_wave_discrete_pos = np.exp(-x_discrete_pos) # Using e^-x for x>=0 as in previous plots

# Plot discrete e^x wave with both cases
plt.figure(figsize=(6, 4))
plt.stem(x_discrete_neg, exp_wave_discrete_neg, label='e^x Wave (x < 0)', basefmt=' ')
plt.stem(x_discrete_pos, exp_wave_discrete_pos, label='e^-x Wave (x >= 0)', basefmt=' ')
plt.title('Discrete e^|x| Wave')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete ramp function
n_ramp = np.arange(-5, 11)
ramp_function = np.maximum(0, n_ramp)

# Plot discrete ramp function
plt.figure(figsize=(6, 4))
plt.stem(n_ramp, ramp_function, label='Discrete Ramp Function', basefmt=' ')
plt.title('Discrete Ramp Function')
plt.xlabel('n')
plt.xticks(n_ramp) # Add this line to set x-axis ticks to all integer values in n_step
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete unit impulse function
n_impulse = np.arange(-5, 6)
unit_impulse = (n_impulse == 0).astype(int)

# Plot discrete unit impulse function
plt.figure(figsize=(6, 4))
plt.stem(n_impulse, unit_impulse, label='Discrete Unit Impulse Function', basefmt='')
plt.title('Discrete Unit Impulse Function')
plt.xlabel('n')
plt.xticks(n_impulse) # Add this line to set x-axis ticks to all integer values in n_step
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for discrete unit step function
n_step = np.arange(-5, 11)
unit_step = (n_step >= 0).astype(int)

# Plot discrete unit step function
plt.figure(figsize=(6, 4))
plt.stem(n_step, unit_step, label='Discrete Unit Step Function', basefmt=' ')
plt.title('Discrete Unit Step Function')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.xticks(n_step) # Add this line to set x-axis ticks to all integer values in n_step
plt.legend()
plt.grid(True)
plt.show()
```
